```python
print(f'******** Importing modules *********')
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Housing price dataset, Square Footage, num of bedrooms as input and Price as output
X = np.array([[2000, 3], [3000, 4], [4000, 3], [5000, 5]])
Y = np.array([250000, 350000, 450000, 550000])
print(f'X shape: {X.shape}, Y shape: {Y.shape}')
```

```python
# Initialize a linear regression model
model = LinearRegression()

model.fit(X, Y)
```

```python
# Output the intercept and coefficients of the trained model
print(f'Intercept: {model.intercept_}')
print(f'Coefficients: {model.coef_}')
```

# Interpretation
Intercept represent the house price when the sq foot is 0 

coefficient[0] - price per sq feet 

coefficient[1] - number of bedrooms has no effect 

- Use the coefficents and intercept to analyze which feature is contributing or not

```python
# Predict the price of a new house with 2500 square feet and 4 bedrooms
new_house = np.array([[3500, 6]])
predicted_price = model.predict(new_house)
print(f'Predicted price: {predicted_price[0]:.2f}')
```

```python
# Plot the data and the regression line
plt.scatter(X[:, 0], Y, color='blue', label='Training data')
plt.plot(X[:, 0], model.predict(X), color='red', label='Regression line')
plt.scatter(new_house[0][0], predicted_price[0], color='green', label='Predicted price')
plt.legend()
plt.show()
```
